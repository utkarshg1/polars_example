# Loan EDA Example


```python
import pandas as pd
```

```python
df = pd.read_csv("Loan_Risk.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
a = df.groupby(by="STATE").agg({"Income": "median", "Age": "median"})
a
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 15))
sns.scatterplot(data=a, x="Age", y="Income", hue="STATE")
plt.show()
```

```python
b = df.groupby(by=["STATE", "Risk"]).agg({"Income": "median", "Age": "median"})
b
```

```python
sns.scatterplot(data=b, x="Age", y="Income", hue="Risk")
```

```python
def scatterplot_group(df: pd.DataFrame, col1: str, col2: str):
    c = df.groupby(by=[col1, col2]).agg({"Income": "median", "Age": "median"})
    plt.figure(figsize=(12, 4))
    sns.scatterplot(data=c, x="Age", y="Income", hue=col2)
    plt.title(f"Median Age vs Income by {col1} and {col2}")
    plt.show()
```

```python
df.columns
```

```python
scatterplot_group(df, "CITY", "Risk")
```

```python
scatterplot_group(df, "CITY", "Married/Single")
```

```python
scatterplot_group(df, "Profession", "Risk")
```
